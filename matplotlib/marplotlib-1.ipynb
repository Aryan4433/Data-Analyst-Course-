```python
# adding label
y=np.array([35,25,25,15])
fruits=['Apple','banana ','grapes','cherries']

plt.pie(y, labels=fruits ,startangle= 90)
plt.show()
```

```python
# explode data
y=np.array([35,25,25,15])
fruits=['Apple','banana ','grapes','cherries']
myexplode=[1,2,1,2]

plt.pie(y, labels=fruits , explode= myexplode)
plt.show()
```

```python
# shadow
y=np.array([35,25,25,15])
fruits=['Apple','banana ','grapes','cherries']
myexplode=[3,3,4,4]

plt.pie(y, labels=fruits , explode= myexplode, shadow=True)
plt.show()
```

```python
# color changing
y=np.array([35,25,25,15])
fruits=['Apple','banana ','grapes','cherries']
col=['red','lightblue','yellow','green']


plt.pie(y, labels=fruits , colors= col)
plt.show()
```

# BOX PLOT

```python
# used to identify outliers
# inter quartile range
# IQR = 75% - 25%
# upper limit = q75+1.5*(IQR)
# lowerlimit = q25- 1.5*(IQR)
```

```python
df= sns.load_dataset('tips')
```

```python
df
```

```python
plt.boxplot(df['tip'])
plt.show()
```

```python
# skewness
df['tip'].skew()
```

```python
df['tip'].describe()
```

```python
q25=df['tip'].quantile(0.25)
q25

q75=df['tip'].quantile(0.75)
q75

IQR= q75-q25
IQR
```

```python
q75=df['tip'].quantile(0.75)
q75
```

```python
IQR= q75-q25
IQR
```

```python
Upper_limit=q75+1.5*(IQR)
Upper_limit
```

```python
Upper_limit
```

```python
Lower_limit=q25-1.5*(IQR)
Lower_limit
```

```python
Lower_limit
```

```python
df[df['tip']>Upper_limit]
```

```python
df[df['tip']<Lower_limit]
```

```python
# Now clear the otliers
# CAPPING
```

```python
new= df.copy()
new['tip']= np.where(new['tip'] > Upper_limit, Upper_limit, np.where(new['tip'] < Lower_limit, Lower_limit, new['tip']))
```
