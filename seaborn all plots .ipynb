```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
# scatterplot: works on both numeric also one num or one cat
```

```python
sns.scatterplot(df['total_bill'],df['tip'],marker="*", hue=df['sex'],data=df)
```

```python
sns.catplot(x='day',y='total_bill',marker="*",data=df)
```

```python
# swarnplot() : works on one cat or one num
```

```python
sns.swarmplot(df['sex'],df['total_bill'],marker="*", hue=df['sex'],data=df)
```

```python
# barplot() works on one num or one cat
```

```python
sns.barplot(df['day'],df['total_bill'],palette=['red','blue'],hue=df['sex'],data=df)
```

```python
plt.bar(df['day'],df['total_bill'],color=['pink','lightblue'],data=df)
```

```python
# lineplot() works on both numeric also one num or one cat
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
sns.lineplot(df['day'],df['total_bill'],color='red',data=df)
# diff btw color and palette
plt.subplot(1,2,2)
sns.lineplot(df['day'],df['total_bill'],palette='husl',hue='sex',data=df);
```

```python
# countplot(): works on one categorical columns
```

```python
sns.countplot(df['sex'],palette=['red','skyblue'],hue='smoker',data=df) # cat
```

```python
# pointplot(): is just like barplot  works on on num or one cat columns
```

```python
sns.pointplot(df['sex'],df['total_bill'])
```

```python
df.groupby(['sex'])['total_bill'].agg(['mean','max','min','sum'])
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
sns.barplot(df['sex'],df['total_bill'])  # avergae return krta h
plt.subplot(1,2,2)
plt.bar(df['sex'],df['total_bill'],data=df) # maximum and minimum return krta h
```

```python
# boxplot() works on both  num or on cat column
```

```python
sns.boxplot(df['day'],df['total_bill'],hue=df['sex'])
```

```python
df[['tip','total_bill']].plot(kind='box')
```

```python
# Violin plot()
```

```python
sns.violinplot(df['sex'],df['total_bill'],palette=['lightblue','black'],hue='smoker',data=df)
```

```python
# Histogram : works on only random numeric columns
```

```python
sns.histplot(df['total_bill'],color=['red'],kde=True)
```

```python
# regressionplot(): works similar as scatter plot but add a linear regression model fit
```

```python
sns.regplot(df['tip'],df['total_bill'], color='purple')
```

```python
# distribution plot() : tell about distribution  of onle one random variable
```

```python
sns.distplot(df['total_bill'],kde=True)
```
