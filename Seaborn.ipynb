```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
 tips.shape
```

```python
tips.describe()
```

```python
tips.notna().sum()
```

```python
tips['smoker'].value_counts()
```

```python
tips.groupby(['smoker'])['sex'].value_counts()
```

```python
# relplot is used to find the relationship
sns.relplot(tips['total_bill'],tips['tip'],hue = 'day', data= tips)
plt.show()
```

```python
sns.relplot(tips['total_bill'],tips['tip'],hue = 'day', style='time' ,data= tips)
plt.show()
```

```python
sns.relplot(tips['total_bill'],tips['tip'],hue = 'day', style='sex' ,size= 'size' ,data= tips)
plt.show()
```

```python
# Line plot
```

```python
tips.head()
```

```python
sns.relplot(tips['total_bill'],tips['tip'],hue = 'sex', col='day' , data= tips)
plt.show()
```

```python
# line plot()
sns.relplot(tips['total_bill'],tips['tip'],hue = 'sex', kind='line', col='day' , data= tips)
```

```python
sns.relplot(tips["total_bill"], tips['tip'], hue='sex',kind='line', col='sex', data= tips)
```

# Catagorical Data Plotting
catplot()
boxplot()
swarmplot()

```python
tips.head()
```

```python
# which day have higher bill
sns.catplot(x='day', y='total_bill',data=tips)
```

```python
# scattering na aye
sns.catplot(x='day', y='total_bill',jitter=False ,data=tips)
```

```python
sns.catplot(x='day', y='total_bill',kind = 'swarm', hue= 'sex', data=tips)
```

```python
sns.catplot(x='day', y='total_bill',kind = 'box', data=tips)
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill',kind = 'box',hue='sex', data=tips)
```

```python
# kiss din kiss kiss ne sabse jayada bill diya hai...
# mat bar graph min and max deta hai
# seaborn # mat bar graph average deta h [mean] deta hai
plt.bar(tips['day'], tips['total_bill'])
```

```python
sns.barplot(x='day', y='total_bill', data=tips, ci=None)
```

```python
tips.groupby(['day'])['total_bill'].agg(['sum','mean'])
```

```python
plt.subplot(1,2,1)
sns.barplot(x='day', y='total_bill', data=tips,hue='sex', ci=None)
plt.subplot(1,2,2)
plt.bar(tips['day'], tips['total_bill'])
plt.show()
```

```python
tips.groupby(['day','sex'])['total_bill'].agg(['max','mean'])
```
