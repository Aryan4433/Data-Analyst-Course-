```python
# data manipulation
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'Age':[400,np.nan,14,np.nan,26,12,13,24,12,23],'Height':[130,np.nan,133,145,123,np.nan,np.nan,np.nan,120,121],'Gender':['Male',np.nan,np.nan,'Female','Female','Female','Male',np.nan,'Male','Male']})
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()/df.shape[0]*(100)
```

```python
# thik kb aur kaise kre data ko mean mode aor median se
# mean outliers nhi h toh
# mode
# median outliers h ton
# outliers
```

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
# check outliers.....
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
df["Height"].mean()              
```

```python
df["Height"].median()    
```

```python
df['Height']=df['Height'].fillna(df["Height"].mean())
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Gender'].mode()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df
```
