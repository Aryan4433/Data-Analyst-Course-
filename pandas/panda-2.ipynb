```python
df=pd.DataFrame({'Age':[400,np.nan,14,np.nan,26,12,13,24,12,23],'Height':[130,np.nan,133,145,123,np.nan,np.nan,np.nan,120,121],'Gender':['Male',np.nan,np.nan,'Female','Female','Female','Male',np.nan,'Male','Male']})
```

```python
# percentage of null values...
df.isnull().sum()/df.shape[0]*(100)
```

```python
# thik kb aur kaise kre data ko mean mode aor median se
# Use mean...outliers nhi hote h
# mode
# Use median... outliers hote h
# outliers
# jb mean aur medain me jada difference ho tb outliers hote h
# jb mean aur medain me jada difference na ho tb outliers nhi hote h
```

```python
df.mean()
```

```python
df.median()
```

```python
# fill age null values
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
# fill height null values
df['Height']=df['Height'].fillna(df['Height'].mean())
```

```python
df
```

```python
# fill null value of gender which is object using Mode...
# mode tell which is occur multiple times...
```

```python
df['Gender'].mode()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df
```

```python
# CONCAT MULTIPLE DATAFRAME
```

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
             'B':['B0','B1','B2','B3'],
             'C':['C0','C1','C2','C3'],
             'D':['D0','D1','D2','D3']},index=[0,1,2,3])




df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
             'B':['B4','B5','B6','B7'],
             'C':['C4','C5','C6','C7'],
             'D':['D4','D5','D6','D7']},index=[4,5,6,7])


df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
             'B':['B8','B9','B10','B11'],
             'C':['C8','C9','C10','C11'],
             'D':['D8','D9','D10','D11']},index=[8,9,10,11])


df4=pd.DataFrame({'E':['E0','E1','E2','E3'],
             'F':['F0','F1','F2','F3'],
             'G':['G0','G1','G2','G3'],
             'H':['H0','H1','H2','H3']})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df4
```

```python
# coloumns same  concat according to row axis=0
# rows same  concat according to coloumn axis=1
```

```python
df=pd.concat([df1,df2,df3],axis=0)
df
```

```python
df=pd.concat([df1,df4],axis=1)
df
```

```python
x=pd.DataFrame({'Key':['k0','k8','k2','k3'],
      'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3']})
```

```python
x
```

```python
y=pd.DataFrame({'Key':['k0','k1','k2','k3'],
       'C':['C0','C1','C2','C3'],
       'D':['D0','D1','D2','D3']})
```

```python
y
```

```python
pd.merge(x,y, how='right', on ='Key')
```

```python
pd.merge(x,y, how='left', on ='Key')
```

```python
pd.merge(x,y, how='outer', on ='Key')
```

```python
pd.merge(x,y, how='inner', on ='Key')
```
